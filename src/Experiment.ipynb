```python
import pandas as pd

df = pd.read_csv('../data/survey_data.csv')
df
```

```python
df.fillna(0,inplace=True)
```

```python
I_fresh  = ((df["근속년수"] >= 0)  & (df["근속년수"] < 3)).astype(int)
I_junior = ((df["근속년수"] >= 3)  & (df["근속년수"] < 10)).astype(int)
I_middle = ((df["근속년수"] >= 10) & (df["근속년수"] < 20)).astype(int)
I_senior = (df["근속년수"] >= 20).astype(int)

df = df.drop(columns='근속년수')
df["근속년수_구분"] = 1 * I_fresh + 2 * I_junior + 3 * I_middle + 4 * I_senior

df
```

```python
CL_map = {'CL1' : 1 , 'CL2' : 2 ,  'CL3' : 3, 'CL4':4}
degree_map = {'고졸' : 1, '학사':2, '석사':2,'박사':3}

df['직급'] = df['직급'].map(CL_map)
df['학력'] = df['학력'].map(degree_map)
df
```

```python
group_map = {'A그룹' :1, 'B그룹' :2, 'C그룹' :3,'D그룹' :4,'E그룹' :5,'F그룹' :6,'G그룹' :7,'H그룹' :8}
df_group = pd.DataFrame(df['소속'].map(group_map),columns=['소속'])
df = pd.get_dummies(df, columns=['소속'],drop_first=False)
df
```

```python
status_map = {'의사 있음':1, '의사 없음':0}

df['이직의사'] = df['이직의사'].map(status_map)
df
```

```python
df.columns
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

TRAIN_FEATURE_SVM = ['복지','회의','교육','업무','상사','동료']

df_selection = pd.DataFrame()
for item in TRAIN_FEATURE_SVM:
    df_selection[item] = df[f'조직문화평가_{item}']

df_selection['회의'] = df['1Q_업무몰입도평가_회의'] + df['2Q_업무몰입도평가_회의']
df_selection['업무'] = df['1Q_업무몰입도평가_고객'] + df['2Q_업무몰입도평가_고객'] + df['1Q_업무몰입도평가_기술력'] + df['2Q_업무몰입도평가_기술력']
df_selection['상사'] = df['1Q_업무몰입도평가_보고'] + df['2Q_업무몰입도평가_보고'] + df['1Q_업무몰입도평가_소통'] + df['2Q_업무몰입도평가_소통']
```

```python
X = df_selection[['복지', '회의', '교육', '업무', '상사', '동료']]
y = df['이직의사']

X_tn, X_te, y_tn, y_te = train_test_split(X,y,random_state=0)

std = StandardScaler()
std.fit(X_tn)

X_tn_std = std.transform(X_tn)
X_te_std = std.transform(X_te)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier

tree_model = DecisionTreeClassifier(random_state=0,max_features=3,max_depth=3)

logistic = LogisticRegression()
svm = SVC(kernel='linear',C=1, degree=4)
GNB = GaussianNB()

voting = VotingClassifier(
    estimators=[
    ('lm',logistic),
     ('gnb',GNB),
     ('svm',svm)
     ],
     voting='hard',
     weights=[1,1,1]
)

rfc = RandomForestClassifier()

rfc.fit(X_tn_std,y_tn)
voting.fit(X_tn_std,y_tn)
tree_model.fit(X_tn_std,y_tn)

y_pred_rfc = rfc.predict(X_te_std)
y_pred_voting = voting.predict(X_te_std)
y_pred_tree = tree_model.predict(X_te_std)

print(y_pred_voting)
print(y_pred_tree)

## 현재 가장 궁금증인, 여러가지 설문조사 데이터를 기반으로한 팀 전체의 가장 큰 이직 결정 요소는 무엇인지 탐색한다.
```

```python
from sklearn.metrics import accuracy_score,classification_report
## 수행결과, 회의, 업무, 상사의 항목이 우리팀의 가장 큰 문제로 지목되었다.
print(rfc.feature_importances_)
report = classification_report(y_te, y_pred_rfc)
print('========RandomForest===========')
print(report)


report = classification_report(y_te, y_pred_voting)
print('========Voting===========')
print(report)

report = classification_report(y_te, y_pred_tree)
print('========tree===========')
print(report)
```

```python
df_personal_regression = df[['직급','학력','근속년수_구분', '소속_A그룹',
       '소속_B그룹', '소속_C그룹', '소속_D그룹', '소속_E그룹', '소속_F그룹', '소속_G그룹', '소속_H그룹']]
```

```python
X = df_personal_regression
y = df['통합점수_전년대비증감']

X_tn, X_te, y_tn, y_te = train_test_split(X,y,random_state=0)


std = StandardScaler()
std.fit(X_tn)

X_tn_std = std.transform(X_tn)
X_te_std = std.transform(X_te)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

svm = SVR(kernel='poly')

svm.fit(X_tn_std,y_tn)
y_pred = svm.predict(X_te_std)

mse = mean_squared_error(y_te,y_pred) #identitiy를 알면 작년 점수를 유추할수 있다. 즉, 그룹 소속의 직급에 따른 
print(mse)
```

```python
## 그렇다면, 여태까지의 실험의 결과를 바탕으로 아래 두가지 실험을 추가적으로 더 수행해보자.
## 1. 이직 요소를 결정하는 요소들의 차원축소를 수행한다. Identitiy에 대한 차원 축소를 진행한다. 그렇게 하여 이직자에 대한 분포를 수행한다.
## 2. 또한 위 차원축소를 바탕을 보여진 분포를 기준으로 그룹별 분포를 수행하여본다. 비지도학습과 실제 label값을 비교하여본다.

import matplotlib.pyplot as plt


pca_identitiy = PCA(n_components=1)
pca_identitiy.fit(X_tn_std)

X_tn_identity_pca = pca_identitiy.transform(X_tn_std)
X_tn_pca_df['pca_comp2'] = X_tn_identity_pca


X_tn_pca_df['target'] = df_group.loc[X_tn.index, '소속'].values
X_tn_pca_df

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))
plt.scatter(X_tn_pca_df['pca_comp1'],
            X_tn_pca_df['pca_comp2'],
            marker='o')
plt.xlabel('Survey_Result')
plt.ylabel('Identitiy')
plt.show()
```

```python
df_new = X_tn_pca_df

X_1 = {}
X_2 = {}
markers = [" ","*","^","o","+","D","s","1","x"]
plt.figure(figsize=(15,8))
for target_i in range(1, 9):
    df_i = df_new[df_new['target'] == target_i]

    X_1[target_i] = df_i['pca_comp1']
    X_2[target_i] = df_i['pca_comp2']

    plt.scatter(X_1[target_i], X_2[target_i], marker=markers[target_i], label=f'target_{target_i}')

plt.legend()
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis




lda = LinearDiscriminantAnalysis(n_components=1)
lda.fit(X_tn_std, y_tn)

X_tn_lda = lda.transform(X_tn_std)
X_te_lda = lda.transform(X_te_std)


df_new = X_tn_pca_df

X_1 = {}
X_2 = {}
markers = [" ","*","^","o","+","D","s","1","x"]
plt.figure(figsize=(15,8))
for target_i in range(1, 9):
    df_i = df_new[df_new['target'] == target_i]

    X_1[target_i] = df_i['pca_comp1']
    X_2[target_i] = df_i['lda_comp']

    plt.scatter(X_1[target_i], X_2[target_i], marker=markers[target_i], label=f'target_{target_i}')

plt.xlabel('PCA_comp')
plt.ylabel('LDA_comp')
plt.legend()
plt.show()
```
